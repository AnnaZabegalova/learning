1. Подготовка базовой модели

Обучите простую модель классификации с помощью DecisionTreeClassifier на данных из датасета vehicles_dataset_prepared.csv. Для этого сделайте шаги:

1. Обучите модель дерева решений с зафиксированным random_state на тренировочной выборке.
2. Сделайте предикт на тестовой выборке.
3. Замерьте точность на тестовой выборке и выведите матрицу ошибок. 
4. Удалите фичи с нулевыми весами по feature_importance из тренировочной и тестовой выборок.
5. Заново обучите модель и измерьте качество.

````python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score, confusion_matrix 
````

````python
df = pd.read_csv('C:/Users/moroz/Documents/zab/ds-intro/22_machine_learning/homework/data/vehicles_dataset_prepared.csv')

df_prepared = df.copy()
df_prepared = df_prepared.drop(['price', 'odometer/price_std'], axis=1)

x = df_prepared.drop(['price_category'], axis=1)
y = df_prepared['price_category']

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=42
)
````

````python
df_prepared.price_category
````

````python
tree = DecisionTreeClassifier()
````

````python
tree.fit(x_train,y_train)
````

````python
predicted_train = tree.predict(x_train)
predicted_test = tree.predict(x_test)
````

````python
predicted_train
````

````python
predicted_test
````

````python
accuracy_score(y_train, predicted_train), accuracy_score(y_test, predicted_test)
````

````python
confusion_matrix(y_test, predicted_test)
````

````python
tree.feature_importances_
tree_imp = list(zip(x.columns, tree.feature_importances_))
tree_imp.sort(key = lambda x: x[1])
tree_imp
````

````python
x2 = x.drop([i[0] for i in tree_imp if i[1] == 0], axis=1)
x2
````

````python
x2_train, x2_test, y_train, y_test = train_test_split(x2, y, test_size=0.3, random_state=42)
````

````python
tree.fit(x2_train,y_train)
````

````python
predicted_train1 = tree.predict(x2_train)
predicted_test1 = tree.predict(x2_test)
predicted_train1, predicted_test1
````

````python
accuracy_score(y_train, predicted_train1), accuracy_score(y_test, predicted_test1)
````

````python
confusion_matrix(y_test, predicted_test1)
````

2. Подготовка модели случайного леса

Обучите простую модель классификации с помощью RandomForestClassifier. Для этого на новых урезанных семплах тренировочной и тестовой выборок обучите модель случайного леса с зафиксированным random_state=50.
